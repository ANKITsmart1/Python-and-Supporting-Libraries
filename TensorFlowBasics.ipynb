```
import tensorflow as tf
```

```
#creating a constant object in tensorflow

hello = tf.constant('Hello World')
```

```
type(hello)
```

```
x = tf.constant(100)
```

```
print(x)
```

```
sess = tf.Session() # Creating a session in Tensorflow 
```

```
#Using a session object, we can run the tensor objects we created above using the run() method.
sess.run(hello)
```

```
# Operations in Tensor Flow

x = tf.constant(2)

y = tf.constant(3)

with tf.Session() as sess:
  print('Operations with Constants:')
  print('Addition:',sess.run(x+y))
  print('Subtraction:',sess.run(x-y))
  print('Multiplication:',sess.run(x*y))
  print('Division:',sess.run(x/y))
```

```
# Advanced Tensorflow Operations (Using Placeholders)

x = tf.placeholder(tf.int32)
y = tf.placeholder(tf.int32)

add = tf.add(x,y)


div = tf.div(x,y)
```

```
with tf.Session() as sess:
  print('Operations using Placeholders')
  print('Addition',sess.run(add,feed_dict={x:20,y:30}))
  print('Division',sess.run(div,feed_dict={x:200,y:30}))
```
