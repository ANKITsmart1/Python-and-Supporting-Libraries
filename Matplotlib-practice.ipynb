```
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
import numpy as np
x = np.linspace(0,2,10)
y = x**2
```

```
x
```

```
y
```

```
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(x,y,'g-')
plt.show()  # Functional way of plotting graphs
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b-') #Used to plot multiple plots at the same time.
```

```
# Object-Oriented Approach to Plotting graphs

fig = plt.figure() # Object declared for plotting graphs

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

axes.plot(x,y)
```

```
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])


axes1.plot(x,y,'b-')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'g-')
axes2.set_title('Smaller Plot')

plt.show()
```

```
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)

for current_ax in axes:
  current_ax.plot(x,y)
  
# axes[0].plot(x,y) - We can also do indexing like this.
```
