                                        Python GUI with iPyWidgets

```python
from ipywidgets import interact,interactive,fixed
```

```python
import ipywidgets as widgets
```

```python
def func(x):
    return x
```

```python
interact(func,x=True)
```

```python
interact(func,x="Hello")
```

```python
@interact(x=True,y=fixed(1.0))
def h(x,y):
    return (x,y)
```

```python
def func(x):
    return x
```

```python
interact(func,x=widgets.IntSlider(min=-100,max=100,step=1,value=0)) #Custom interact() function
```

```python
interact(func,x=(-100,100,1)) #Abbreviated custom interact() function
```

```python
@interact(x=(0.0,20.0,0.5))
def h(x=5.0):
    return x
```

```python
interact(func,x=['hello','hi','heya'])
```

```python
interact(func,x={'hello':9,'hi':90,'heya':7})
```

```python
from IPython.display import display
```

                                                    Widget Basics

```python
import ipywidgets as widgets
```

```python
w = widgets.IntSlider()
```

```python
from IPython.display import display
```

```python
display(w)
```

```python
display(w)
```

```python
w.close() #closes all instances of the active widget
```

```python
w = widgets.IntSlider()
```

```python
display(w)
```

```python
w.value # Gives us the current value from the running widget
```

```python
w.keys # We can obtain a list of keys associated with the widget and then manipulate the widget accordingly.
```

```python
a = widgets.FloatText()
b = widgets.FloatSlider()

display(a,b)

mylink = widgets.jslink((a,'value'),(b,'value'))
```

```python
a = widgets.FloatText()
b = widgets.FloatSlider()

display(a,b)

mylink = widgets.jslink((a,'value'),(b,'max'))
```

```python
mylink.unlink() # Removes all the links between the values
```
