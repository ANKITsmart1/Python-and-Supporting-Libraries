```
import seaborn as sns
```

```
%matplotlib inline
```

```
tips = sns.load_dataset('tips')
```

```
tips.head()
```

# Distribution Plots

```
sns.distplot(tips['total_bill'],kde=False) # Gives a histogram as the output.
```

```
sns.jointplot(x='total_bill',y='tip',data=tips) # Used to combine two Uni-variate distributions in one attempt.
```

```
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') # Kind Parameter is used to enhance visualizations within the joint-plot distribution.
```

```
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```
sns.pairplot(tips) # Paiplot helps to visualize literally all possible combinations of the numerical columns.
```

```
sns.pairplot(tips,hue='sex') # The argument "Hue" specifies the categorical column values within the Pairplot distribution.
```

```
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```
sns.rugplot(tips['tip'])
```

# Categorical Plots

```
sns.barplot(x='sex',y='total_bill',data=tips) # Gives a bar graph representation of the relation between a categorical and a non-categorical column.

                                              # we can have an argument "estimator" to perform different aggregrate operations on the columns mentioned here. By default, we perform Mean.
```

```
sns.countplot(x='sex',data=tips) # Countplot simply counts the number of occurrences of the column values.
```

```
sns.boxplot(x='day',y='total_bill',data=tips) # Used to represent quartile information regarding the featured data
```

```
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```
sns.violinplot(x='day',y='total_bill',data=tips)
```

```
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```
sns.stripplot(x='day',y='total_bill',data=tips)
```

```
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```
sns.factorplot(x='day',y='total_bill',data=tips,kind='strip')
```

# Matrix Plots

```
import seaborn as sns
```

# *HeatMaps*

```
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```
flights.head()
```

```
# The need for matrix plots arises from the fact that most of the time,
# the rows of the dataset are not having actual categorical values as 
# the columns and it becomes difficult to co-relate between them.
```

```
tc = tips.corr()
```

```
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```
flights
```

```
ft = flights.pivot_table(index='month',columns='year',values='passengers')
ft
```

```
sns.heatmap(ft,cmap='magma',linecolor='white',linewidths=0.5)
```

# *ClusterMaps*

```
sns.clustermap(ft, cmap='coolwarm',standard_scale=1)
```

# Grids

```
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```
iris['species'].unique()
```

```
sns.pairplot(iris,hue='species',palette='magma')
```

```
import matplotlib.pyplot as plt
g = sns.PairGrid(iris) # PairGrid allows us to showcase different plots within a grid.
g.map_diag(sns.distplot) 
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```
tips = sns.load_dataset('tips')
```

```
tips.head()
```

```
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'tip','total_bill')
```

# Regression Plots

```
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',palette='magma',markers=['v','o'],scatter_kws={'s':80})
```

```
sns.lmplot(data=tips,x='total_bill',y='tip',col='sex')
```

```
sns.lmplot(data=tips,x='total_bill',y='tip',col='day',hue='sex',palette='pastel',aspect=0.5,size=8)
```
