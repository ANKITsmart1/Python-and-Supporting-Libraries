                                                   Generators in Python

```python
#Generators are special functions which doesn't just return a value and get finished.

#They can resume opeartion from the point when they generated the last return value. Yield keyword is used for this purpose.

#It is more memory-efficient( this is completely intuitional! )

def create_cubes(n):
    for x in range(n):
        yield x**3
        
        
for x in create_cubes(9):
    print(x)

```

```python
def gen_fibon(n):
    a = 1
    b = 1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
for number in gen_fibon(10):
    print(number)
```
